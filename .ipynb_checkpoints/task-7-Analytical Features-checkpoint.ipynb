```python
covid_data['DEATHS'] = covid_data['DEATHS'].fillna(covid_data['DEATHS'].mean())
```

```python
# Stationarity is a crucial assumption in time 
# series analysis. A time series is stationary if its statistical properties
# (mean, variance) are constant over time. Non-stationary data can lead to misleading models.
# You can check for stationarity using statistical tests like the Augmented Dickey-Fuller (ADF) test:

from statsmodels.tsa.stattools import adfuller

result = adfuller(covid_data['DEATHS'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

# Interpretation
if result[1] > 0.05:
    print("Series is not stationary")
else:
    print("Series is stationary")
```

```python
 # Parameter Selection (p, d, q)   in ARIMA
# no need to chage order differencing 
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


plot_acf(covid_data['DEATHS'])
plot_pacf(covid_data['DEATHS'])
plt.show()
```

```python
# From above plot we can aslo say that series in stationary
# The d parameter in an ARIMA model represents the order of differencing needed to make the series stationary. 
# d= 0 because that the series is already stationary and does not require differencing
# The PACF plot shows a significant spike at lag 1, then cuts off, possible p=1
# q (MA terms): Look at the ACF plot (the top one) to determine the number of moving average terms.
# The q parameter corresponds to the last significant lag after which most other lags are insignificant.
# there are significant lags at least up to lag 2,q=2
```
